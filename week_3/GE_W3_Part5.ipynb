```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
starbucks = pd.read_csv("./data/starbucks/starbucks.csv")
pima_data = pd.read_csv("./data/pima/diabetes.csv")

data_pima = pima_data[pima_data["BloodPressure"]>0]
data_pima = data_pima[data_pima["BMI"]>0]

data_starbucks = starbucks[starbucks["Calories"]>0]
data_starbucks = data_starbucks[data_starbucks[" Protein (g) "]>0]
```

```python
# Subplot Grid
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# (a) Top-left: Histogram of BloodPressure
sns.histplot(
    data=data_pima, 
    x='BloodPressure', 
    kde=False, 
    ax=axes[0, 0], 
    color='skyblue')
axes[0, 0].set_title('Histogram of Blood Pressure (PIMA)')
axes[0, 0].set_xlabel('Blood Pressure')
axes[0, 0].set_ylabel('Frequency')

# (b) Top-right: Scatter plot of Calories vs Protein (g) from Starbucks
sns.scatterplot(
    data=data_starbucks, 
    x='Calories', 
    y=' Protein (g) ', 
    ax=axes[0, 1], 
    color='green')
axes[0, 1].set_title('Calories vs Protein (Starbucks)')
axes[0, 1].set_xlabel('Calories')
axes[0, 1].set_ylabel('Protein (g)')


# (c) Bottom-left: Count plot of Beverage Category
sns.countplot(
    data=data_starbucks, 
    y='Beverage_category', 
    order=data_starbucks['Beverage_category'].value_counts().index, 
    ax=axes[1, 0], 
    palette='Set2')
axes[1, 0].set_title('Beverage Category Count (Starbucks)')
axes[1, 0].set_xlabel('Frequency')
axes[1, 0].set_ylabel('Beverage Category')

# (d) Bottom-right: Box plot of Age grouped by Outcome
sns.boxplot(
    data=data_pima,
    x='Outcome', 
    y='Age', 
    ax=axes[1, 1], 
    palette='coolwarm')
axes[1, 1].set_title('Age by Diabetes Outcome (PIMA)')
axes[1, 1].set_xticklabels(['0', '1'])
axes[1, 1].set_ylabel('Ages')

fig.suptitle("Subplots of PIMA and Starbucks")
plt.tight_layout()
plt.show()
```
