```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Loading
starbucks = pd.read_csv("./data/starbucks/starbucks.csv")
```

```python
# Univariate Analysis
```

```python
# Histogram
fig_mpl_hits, ax_mlp_hits = plt.subplots(figsize=(5,3))
# Age Data
ax_mlp_hits.hist(
    starbucks['Calories'].values,
    bins=15,
    color="skyblue",
    edgecolor="black"
)
ax_mlp_hits.set_title("Calories distribution of Starbucks dataset")
ax_mlp_hits.set_xlabel("Calories")
ax_mlp_hits.set_ylabel("Number of Items")
plt.show()
```

```python
# Boxplot
fig_mpl_box, ax_mlp_box = plt.subplots(figsize=(5,3))
ax_mlp_box.boxplot(
    starbucks['Calories'].values
)
ax_mlp_box.set_title("Calories Distribution of Starbucks Dataset")
ax_mlp_box.set_ylabel("Calories")
ax_mlp_box.set_xticks([])
plt.show()
```

```python
# Interpretation
# 50% of all items have less than 200kCal of calories.
# The most number of items as per the modal class at 200kCal on the historgram reveals the highest number of items (40 items) have around the median number of calories (185 kCal).
# 75% of all items are below 260 kCal with a spread of 260kCal, the upper quartile (Q3) has a large spread of data which in itself is 250kCal.
# Calories are distributed unevenly, as the data is positively skewed.
```

```python
# Bivariate Analysis
```

```python
data_starbucks = starbucks[starbucks["Calories"]>0]
data_starbucks = data_starbucks[data_starbucks[" Sugars (g)"]>0]
```

```python
# Scatterplot
fig, ax=plt.subplots(figsize=(5,3))
sns.scatterplot(
   data=data_starbucks,
    x="Calories",
    y=" Sugars (g)",
    ax=ax,
    color="#29429AD6" 
)
ax.set_title("Scatter Plot between Calories and Sugar in Starbucks Food Items")
ax.set_xlabel("Calories")
ax.set_ylabel("Sugar (g)")
plt.show()
```

```python
# Interpretation
# There is an apparent positive correlation between sugar and calories. 
# As sugar increases, so does the calories in a food item.
```
