```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading
pima_data = pd.read_csv("./data/pima/diabetes.csv")
```

```python
pima_data.head()
```

```python
pima_data.info()
```

```python
pima_data.describe()
```

```python
# Univariate Analysis
```

```python
# KDE Plot
fig_sns_hits, ax_sns_hits = plt.subplots(figsize=(5,3))
sns.kdeplot(
    data=pima_data,
    x="Age",
    color='purple',
    ax=ax_sns_hits
)
ax_sns_hits.set_title("Age distribution (KDE) of PIMA dataset")
ax_sns_hits.set_xlabel("Ages")
ax_sns_hits.set_ylabel("Frequency")
plt.show()
```

```python
# Histogram with KDE Plot
fig_sns_hits, ax_sns_hits = plt.subplots(figsize=(5,3))
sns.histplot(
    data=pima_data,
    x="Age",
    bins=15,
    kde=True,
    palette="Pastel2",
    ax=ax_sns_hits
)
ax_sns_hits.set_title("Age distribution (with KDE) of PIMA dataset")
ax_sns_hits.set_xlabel("Ages")
ax_sns_hits.set_ylabel("Frequency")
plt.show()
```

```python
# Bivariate Analysis
```

```python
# Side-by-side violin plots
fig_sns_violin, ax_sns_violin = plt.subplots(figsize=(2,4))
sns.violinplot(
    data=pima_data,
    y = 'Glucose',
    x = "Outcome",
    ax=ax_sns_violin,
    order=["0", "1"],
    palette="Pastel2"
)
ax_sns_violin.set_title("Violin Plot between Glucose and Outcome of PIMA dataset")
ax_sns_violin.set_ylabel("Glucose")
plt.show()
```

```python
# Interpretation
# If a person has diabetes they have higher glucose levels on average as compared to a person who doesn't have diabetes.
# 75% of people with diabetes have higher glucose levels than 50% of people without diabetes.
```

```python
# Multivariate Analysis
```

```python
data_pima = pima_data[pima_data["Age"]>0]
data_pima = data_pima[data_pima["BMI"]>0]
data_pima
```

```python
# Scatterplot of BMI and Age
fig, ax=plt.subplots(figsize=(5,3))
sns.scatterplot(
   data=data_pima,
    x="Age",
    y="BMI",
    ax=ax,
    color="#4F6ACBD5" 
)
ax.set_title("Scatter Plot of Age vs. BMI")
ax.set_xlabel("Age")
ax.set_ylabel("BMI")
plt.show()
```

```python
# Scatterplot of BMI, Age, and Outcome
fig, ax=plt.subplots(figsize=(5,3))
sns.scatterplot(
   data=data_pima,
    x="Age",
    y="BMI",
    ax=ax,
    hue="Outcome",
    palette="coolwarm"  
)
ax.set_title("Scatter Plot of Age vs. BMI")
ax.set_xlabel("Age")
ax.set_ylabel("BMI")
plt.legend(title='Outcome')
plt.show()
```

```python
# Scatterplot of BMI, Age, Outcome, and Pregnancies
fig, ax=plt.subplots(figsize=(10, 6))
sns.scatterplot(
    data=data_pima,
    x='Age',
    y='BMI',
    hue='Outcome',
    size='Pregnancies',
    palette='coolwarm',
    sizes=(20, 200),
    alpha=0.7
)

plt.title('BMI vs Age Colored by Diabetes Outcome and Sized by Pregnancies', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('BMI', fontsize=12)
plt.legend(title='Legend')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Interpretation
# Higher BMI tends to be more associated with diabetic individuals (Outcome = 1), especially among those in the middle-age range (3050)
# The size of the points also shows that women with more pregnancies are more likely to have diabetes. 
# However, diabetes appears across a broad age and BMI range, suggesting multiple contributing factors.
```

```python
# Correlation Heatmap
```

```python
selected_features = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'Age']

corr_matrix = data_pima[selected_features].corr()
print(corr_matrix)
```

```python
fig, ax=plt.subplots(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Selected PIMA Features')
plt.show()
```

```python
# Interpretation
# The heatmap shows that Age and Pregnancies have moderate positive correlation.
# The heatmap also shows that Glucose levels and Age also have moderate positive correlation
# BMI is weakly correlated with the rest. 
# These insights help identify potential predictors of diabetes.
```
